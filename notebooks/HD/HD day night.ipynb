```python
# Dir where all the HD files are
base_dir = "/home/parth/Machine Learning/Datasets/North Corp/Ajanta"
```

```python
import os
import pandas as pd
import numpy as np
```

```python
data = []
files = os.listdir(base_dir)
for file in files:
    if file.endswith(".csv"):
        data.append(pd.read_csv(os.path.join(base_dir , file)))
```

```python
for vehicle in data[0].vehicleid.unique():
    print(len(data[0][data[0].vehicleid == vehicle].gtpdriverid.unique()))
```

```python
data[0].columns
```

```python
# ADD columns to be included
useful_cols = [ "vehicleid" , "updateddate" , "total_engine_hours"]
```

```python
for i in range(len(data)):
    data[i] = data[i][useful_cols]
```

```python
for i in range(len(data)):
    data[i]["date"] = data[i]["updateddate"].apply(lambda x: x.split(' ')[0])
```

```python
for i in range(len(data)):
    data[i]["Day_Hours"] = data[i]["updateddate"].apply(lambda x: int(x.split(' ')[1].split(':')[0]))
```

```python
from itertools import chain
night_range = chain(range(6) , range(22 , 24))
nt = []
for n in night_range:
    nt.append(n)
night_range = nt
del nt
day_range = range(6 , 16)
evening_range = range(16 , 22)
```

```python
data
```

```python
complete_data = []
for i in range(len(data)):
    first = True
    for day in data[i].date.unique():
        x = data[i][ np.logical_and( data[i].date == day  ,  data[i].Day_Hours.isin(night_range))]["total_engine_hours"]
        night_drive = x.max() - x.min()
        x = data[i][ np.logical_and(data[i].date == day ,data[i].Day_Hours.isin(day_range))]["total_engine_hours"]
        day_drive = x.max() - x.min()
        x = data[i][np.logical_and(data[i].date == day , data[i].Day_Hours.isin(evening_range))]["total_engine_hours"]
        eve_drive = x.max() - x.min()
        
        mini_dat = pd.DataFrame([[day , night_drive , day_drive , eve_drive]] , 
                                columns=["Day" , "Night Drive" , "Day Drive" , "Evening Drive"])
        if first:
            complete_data.append(mini_dat)
            first = False
        else:
            complete_data[i] = complete_data[i].append(mini_dat)
```
